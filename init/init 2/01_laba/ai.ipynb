### Початкова програма на Python

```python
# Це коментар у Python. Все, що знаходиться після символу #, не впливає на роботу програми.
# Коментарі призначені для пояснення коду.

# Ми почнемо зі створення змінної. Змінна - це контейнер для зберігання даних.
# У нашому випадку, ми створюємо змінну з ім'ям "повідомлення" і присвоюємо їй рядок.

повідомлення = "Привіт, світ!"  # Ми створюємо змінну "повідомлення" і присвоюємо їй рядок.

# Тепер, коли у нас є змінна з текстом, давайте виведемо її на екран.

print(повідомлення)  # Функція "print" виводить текст змінної "повідомлення" на екран.

# Ваша програма завершується. Код вище виведе текст "Привіт, світ!" на екран.
```

### Змінні у Python

```python
# Це коментар у Python. Все, що знаходиться після символу #, не впливає на роботу програми.
# Коментарі призначені для пояснення коду.

# Змінні можуть бути глобальними або локальними. Глобальні змінні доступні в усій програмі, 
# локальні - лише в межах певної функції або блоку коду.

# Глобальна змінна:
глобальна_змінна = "Ця змінна глобальна і доступна в усій програмі"

# Локальна змінна:
def приклад_функції():
    локальна_змінна = "Ця змінна локальна і доступна тільки в межах функції"
    print(локальна_змінна)

# Тепер ми зможемо запросити в користувача ввести значення для змінних.
повідомлення = input("Введіть повідомлення: ")

# Виводимо на екран введене користувачем повідомлення:
print("Ви ввели таке повідомлення:", повідомлення)

# Ваша програма завершується. Код вище дозволить ввести значення для змінної "повідомлення" 
# та виведе це значення на екран.
```

Основні зміни та додатки:

Ми визначили глобальну змінну глобальна_змінна поза будь-якою функцією. Ця змінна доступна в усій програмі.

Ми визначили функцію приклад_функції(), в якій є локальна змінна локальна_змінна. Ця змінна доступна тільки всередині функції приклад_функції() і не може бути використана в інших частинах програми.

Ми використали функцію input() для отримання введеного користувачем значення і зберегли його в змінній повідомлення. input() очікує від користувача введення та повертає введену стрічку (рядок).

Ми вивели введене користувачем повідомлення на екран.

За допомогою цієї програми користувач може ввести своє повідомлення, і воно буде виведено на екрані. Однак слід звернути увагу, що введені дані залишаються рядками (строками), і для числових операцій їх слід буде перетворити в числа, якщо потрібно.

```python
def my_function()-> str:
    local_variable = "Ця локальна змінна "
    print(local_variable)
    return local_variable


value_from_function = my_function()

print(f"Функція поверне значення: {value_from_function}")
```
