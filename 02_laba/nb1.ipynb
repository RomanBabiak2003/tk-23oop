- вбудовані типи даних (текстові/стрічкові)

```python
print("Це текстові дані які будуть виведені")
a = "Ця змінна містить текстові дані"
print(a)
b = str("Це буде текстові дані через клас str")
print(b)

print("Рахуємо скільки разів зустрічається буква е в об'єкті b: ",b.count("е"))
print("Рахуємо скільки разів зустрічається буква і в текстовій змінній а: ",a.count("і"))
print("Рахуємо скільки разів зустрічається буква і в простій стрічці: ", "Це буде стічка без присвоєння змінній чи створення обєкту".count("і"))
```

* конкатенація(додавання/зчепленя ) та множення стрічок

```python
c = a + "|Це буде рядок зчепленя|" + b
print (c)
d = a.__add__("|Це буде рядок зчепленя|".__add__(b))
print ("В обох випадках будемо мати \n", c, "\n", d)
#ці два записа рівноціні 
e = a * 4
f = a.__mul__(4)
print("Що бдуе при множення стрічки на число? \n", e, "\n", f)
#Найпростіше можна побавитись у вивід символів заповнювачів\
print("~=+=~" * 30)
```

* Числові тип даних

```python
a = 7
b =int(7)
print("Цілі числа автоматично будуть відноситись до класу:", type(a), type(b))
c = 7.7
d = float(a)
print("Якщо число є дробовим то буде тип:", type(c),
       "\nАбо можна явно перетворити", type(a), "в", type(d), 
       "\nПісля перетворення число", a, "тепер буде", d)
```

```python
b += 0.1 #це ще називають накопичуванням
print("В цьому випадку має наєвне перетворення:", type(b), "\nРезультат (буде з похибкою):", b)

e = a + 0.1
f = 0.1 + a
print(
    "\nТут все нормаль працює", type(e), "значення", e,
    "\nТут також результат буде", type(f), "значення", f,
)

e = a.__add__(0.1)
f = 0.1.__add__(a)
print("З використанням методів класу додавання буде працювати тільки за певних умов.",
       "\nДля цілих типів даних можна додавати тільки цілі, інакший результат буде", type(e), "значення ", e,
       "\nЯких ж цілі числа будем додавати з плаваючою крапкою то буде ", type(f),"значення", f,
)
```

```python
1 = [1, 1.2, "text", a]
11  = list(1)


```

```python
t = (1, 1.2, "text", a)
tt = tuple(t)

print("Початкове значення кортиджу:",t)

print("Навіть використовуючи зчеплення, ми не модифікували наш tuple, а просто повернули нове значення:", t.__add__((1,)))

print("Кількість 1 буде:",t.count(1))
print("Знайдемо під яким індексом знаходиться число 1:", t.index(1))
```
