- Запитались в ChatGPT як вивести піраміду з символів

Напиши будь ласка на Python код програми, що виводить піраміду з символів

```python
def draw_pyramid(height):
    for i in range(1, height + 1):
        spaces = " " * (height - i)
        stars = "*" * (2 * i - 1)
        print(spaces + stars)

try:
    height = int(input("Введіть висоту піраміди: "))
    draw_pyramid(height)
except ValueError:
    print("Будь ласка, введіть коректне число.")
```

- робота з циклами
> `while` - умовний цикл ітерації будуть виконуватись поки умова є ітинною (може бути безмежним циклом)

> `for` - із задою кількістю ітерацій

```python
from random import randint
n = ["Roman", "Eli", "Mark"]
h = ["Anime", "Art", "Worker"]

j = 0
while j < 3:
    print(">>>{name} любить {hobbi}".format(name=n[j], hobbi=h[randint(0,2)]))
    j+=1

print("="*10,"Це вже не входить у цикл, цикл закінчився. \nЦикл While не дуже підходить для визначеної кількості ітерацій.")

# навіть для for можна йти по різному
for j in range(len(n)):
     print(">>>{name} любить {hobbi}".format(name=n[j], hobbi=h[randint(0,2)]))

print("="*10,"Це вже виконається поза циклом:")

print(f"Демонструємо що буде виконувати функція range: {list(range(3))}")

# варіант 2 для for\
for j in n:
    print(">>>{name} любить {hobbi}".format(name=j, hobbi=h[randint(0,2)]))

# Спростимо код ще більше
print("="*10)
for j in n:
        print(f">>>{j} любить {h[randint(0,2)]}")
```

```python
attempts = 4 # Кількість ітерацій
required_number = 6 # число яке має випасти на кубику
dice_number = int()
# якщо ми не будемо використовувати нашу ітераційну змінну у циклі, то можна її назвати _ (просто нижнє підкреслення)
for _ in range(attempts):
    dice_number = randint(1,6)
    print(f"Кидаємо кубик, випало значення {dice_number}. Отримане значення по відношенню до бажаного буде: {required_number == dice_number}.")
    if required_number == dice_number:
        print(f"Умова виконана на {_ + 1} ітерації, виходимо з циклу.")
        break
else:
    print("Задане значення не було отримане.")
```

- пробуємо one-liner (використання синтаксису if або for у одному рядку)
> може бути використано для простих умов які можна помістити в один рядок

```python
name = "Romane"

# Кононічний вигляд

if name in n:
    print(f"Імя {name} знаходиться в списку n = {n}")
else:
    print(f"Немає імені {name} у списку, додаємо його: {n.append(name)}")  

# Запис в один рядок

#print(f"Імя {name} знаходиться в списку n = {n}" if name in n else f"Немає імені {name} у списку.")
print(f"Імя {name} знаходиться в списку n = {n}" if name in n else f"Немає імені {name} у списку, додаємо його: {n.append(name)}")

print(f"Кінцевий вигляд списку буде: {n}")
```

```python
pn = list()
# канонічний запис циклу
for name in n:
    len(name)
    pn.append(f"Ім'я {name} має {len(name)} букв")

print(pn)

# переписуємо простий цикл в One liner

pn2 = list([f"Ім'я {name} має {len(name)} букв" for name in n])

print(pn2)
```

```python
from time import time
n = ["Roman", "Eli", "Mark", "Тарас", "Влад", "Тетяна", "Назар"]
print(f"Початковий вигляд списку: {n}")
# варіант використання for i if
start = time()
pn1 = list()
for name in n:
    if len(name) == 5:
        pn1.append(f"Це {name}!")
    else:
        pn1.append(f"Оце так {name}!")
end = time()
print(pn1, f"\nСписок заповнюється {end - start} секунд")
# варіант використання for та if та модифікація за допомогою continue



pn = list()
for name in n:
    if len(name) == 5:
        pn.append(f"Це {name}!")
        continue
    else:
        pn.append(f"Оце так {name}!")

print(pn)
# варіант використання for ta if в одному рядку
pn2 = list([f"Це {name}!" if len(name) == 5 else f"Оце так імя {name}!" for name in n ])
print(pn2)

```
