- Запитались в ChatGPT як вивести піраміду з символів

Напиши будь ласка на Python код програми, що виводить піраміду з символів

```python
def draw_pyramid(height):
    for i in range(1, height + 1):
        spaces = " " * (height - i)
        stars = "*" * (2 * i - 1)
        print(spaces + stars)

try:
    height = int(input("Введіть висоту піраміди: "))
    draw_pyramid(height)
except ValueError:
    print("Будь ласка, введіть коректне число.")
```

- знайомимось з списками та кортеджі
> `List` - впорядкована колекція змінних типів даних;

> `Tuple` -впорядкована колекція незмінних даних;

```python
l = [1, 1.2, "текст", a]
ll = list(l)

print("Початкове значення списку:",l)

l.append("Це кінець") # Вставляємо в кінець
l.insert(1, "Це буде на індексі 1") # Вставляємо на позицію 1
l.pop(-2) # видаляємо 5

print("Після змін, список має вигляд:",l)

l.extend (ll)
print("Якщо зчепити два списки, буде:", l)
```

```python
t = (1, 1.2, "text", a)
tt = tuple(t)

print("Початкове значення кортиджу:",t)

print("Навіть використовуючи зчеплення, ми не модифікували наш tuple, а просто повернули нове значення:", t.__add__((1,)))

print("Кількість 1 буде:",t.count(1))
print("Знайдемо під яким індексом знаходиться число 1:", t.index(1))
```
